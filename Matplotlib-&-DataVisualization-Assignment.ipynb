```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn import decomposition
from sklearn import datasets
%matplotlib inline
```

```python
iris = datasets.load_iris()
```

```python
iris
```

```python
X = iris.data[:, :2]  
y = iris.target
```

```python
fig = plt.figure(1, figsize=(10, 10))
ax = Axes3D(fig)
X_reduced = decomposition.PCA(n_components=3).fit_transform(iris.data)
ax.scatter(X_reduced[:, 0], X_reduced[:, 1], X_reduced[:, 2], c=y)
ax.set_title("IRIS Data")
ax.set_xlabel("1st")
ax.w_xaxis.set_ticklabels([])
ax.set_ylabel("2nd")
ax.w_yaxis.set_ticklabels([])
ax.set_zlabel("3rd")
ax.w_zaxis.set_ticklabels([])
```

## 1. Scipy

```python
import seaborn as sns
```

```python
Max = np.array([39, 41, 43, 47, 49, 51, 45, 38, 37, 29, 27, 25])
Min = np.array([21, 23, 27, 28, 32, 35, 31, 28, 21, 19, 17, 18])
```

```python
months = np.arange(12) #Our Months will help in creating circles of the diagram
days = np.linspace(0,12,num=365) #Our Days will help in creating a smooth line as values are 365....
```

```python
print("Months:\n",months)
print("\nDays:\n",days[:20])# Total 365 data
```

```python
dia=sns.scatterplot(months,Max,color='red')
sns.scatterplot(months,Min,color='blue')
dia.set(xlabel='Months', ylabel='Temperature')
```

```python
from scipy import optimize
def yearly_temps(times, avg, ampl, time_offset):
    return (avg + ampl * np.cos((times + time_offset) * 1.8 * np.pi / times.max()))

res_max, cov_max = optimize.curve_fit(yearly_temps, months,Max, [40, 20, 0])
res_min, cov_min = optimize.curve_fit(yearly_temps, months,Min, [-40, 20, 0])
```

```python
max_tem=yearly_temps(days,*res_max)
min_tem=yearly_temps(days,*res_min)
```

```python
dia_final=sns.scatterplot(months,Max,color='red')
sns.lineplot(days, max_tem, color='red')
sns.scatterplot(months,Min,color='blue')
sns.lineplot(days, min_tem, color='blue')
dia_final.set(xlabel='Months', ylabel='Temperature')
```
